```python
import seaborn as sb

#loading iris dataset

df = sb.load_dataset("iris")
df.head(102)
```

```python
#checking values of species column
df["species"].value_counts()
```

```python
#converting categorical value to numerical
d={'versicolor':2, 'virginica':1,'setosa':0}
df['species']=df['species'].map(d)
```

```python
#value s after conversion
df["species"].value_counts()
```

```python
# target col is the col which we try to predict (species)
y = df['species']
X = df.drop('species',axis=1)
```

```python
#using the decision tree algo
from sklearn.tree import DecisionTreeClassifier
decision_Tree = DecisionTreeClassifier()
decision_Tree.fit(X,y)
```

```python
# #visualizing the graph
#ploting desicision tree of our model
from matplotlib import pyplot as plt
from sklearn.tree import plot_tree
fig = plt.figure(figsize=(20,15))
_ = plot_tree(decision_tree=decision_Tree,feature_names = X.columns,class_names=["setosa","vercicolor","verginica"],filled = True,precision=4,rounded =True)
```

```python
#prediction 
#some values taken for prediction 

# 5.1	3.5 	1.4 	0.2 	setosa
# 5.8	 2.7	5.1 	1.9 	virginica
# 5.1	 2.5	 3.0	1.1 	versicolor



print(decision_Tree.predict([[5.1,3.5,1.4,0.2]]))     # 0 means "setosa species"
print(decision_Tree.predict([[5.8,2.7,5.1,1.9]]))     # 1 means "virginica" species
print(decision_Tree.predict([[5.1,2.5,3.0,1.1]]))     # 2 means "versicolor" species
```
